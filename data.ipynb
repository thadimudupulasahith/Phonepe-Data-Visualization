```python
import os
import json
import pandas as pd

# Set the path to the data folder containing state-level data
data_folder_path = "C:/Users/Sahith/Desktop/Phonepe_Pulse/data/aggregated/transaction/country/india/state/"

# Get the list of states in India
state_list = os.listdir(data_folder_path)

# Initialize an empty dictionary to store the transaction data
transaction_data = {'State': [], 'Year': [], 'Quarter': [], 'Transaction Type': [], 
                    'Transaction Count': [], 'Transaction Amount': []}

# Loop through each state and extract the transaction data
for state in state_list:
    state_folder_path = os.path.join(data_folder_path, state)
    year_list = os.listdir(state_folder_path)
    
    for year in year_list:
        year_folder_path = os.path.join(state_folder_path, year)
        quarter_list = os.listdir(year_folder_path)
        
        for quarter in quarter_list:
            json_file_path = os.path.join(year_folder_path, quarter)
            with open(json_file_path, 'r') as json_file:
                data = json.load(json_file)
                try:
                    for transaction in data['data']['transactionData']:
                        name = transaction['name']
                        count = transaction['paymentInstruments'][0]['count']
                        amount = transaction['paymentInstruments'][0]['amount']
                        
                        transaction_data['State'].append(state)
                        transaction_data['Year'].append(year)
                        transaction_data['Quarter'].append(int(quarter.strip('.json')))
                        transaction_data['Transaction Type'].append(name)
                        transaction_data['Transaction Count'].append(count)
                        transaction_data['Transaction Amount'].append(amount)
                except:
                    pass

# Create a pandas dataframe from the extracted transaction data
transaction_df = pd.DataFrame(transaction_data)
transaction_df.to_csv("transaction_df.csv")
# Print the dataframe
print(transaction_df.head())
```

```python
import os
import json
import pandas as pd

# Set the path to the state-level data
path = "C:/Users/Sahith/Desktop/Phonepe_Pulse/data/aggregated/user/country/india/state/"

# Get a list of all the states
user_state_list = os.listdir(path)

# Initialize an empty dictionary to store the extracted data
data_dict = {
    'State': [],
    'Year': [],
    'Quarter': [],
    'Brand': [],
    'Brand_count': [],
    'Brand_percentage': []
}

# Loop through each state and extract the data
for state in user_state_list:
    # Set the path to the state's data
    state_path = os.path.join(path, state)
    year_list = os.listdir(state_path)
    
    # Loop through each year for the state and extract the data
    for year in year_list:
        year_path = os.path.join(state_path, year)
        quarter_list = os.listdir(year_path)
        
        # Loop through each quarter for the year and extract the data
        for quarter in quarter_list:
            quarter_path = os.path.join(year_path, quarter)
            
            # Load the JSON file and extract the brand data
            with open(quarter_path, 'r') as f:
                data = json.load(f)
                try:
                    for device in data["data"]["usersByDevice"]:
                        brand = device["brand"]
                        brand_count = device["count"]
                        brand_percentage = device["percentage"]
                        
                        # Add the extracted data to the dictionary
                        data_dict['State'].append(state)
                        data_dict['Year'].append(year)
                        data_dict['Quarter'].append(int(quarter.strip('.json')))
                        data_dict['Brand'].append(brand)
                        data_dict['Brand_count'].append(brand_count)
                        data_dict['Brand_percentage'].append(brand_percentage)
                except:
                    pass

# Create a pandas dataframe from the dictionary
user_by_device = pd.DataFrame(data_dict)
user_by_device.to_csv('user_by_device.csv')
# Display the resulting dataframe
user_by_device
```

```python
# Set the path to the directory containing state-wise transaction data
path = "C:/Users/Sahith/Desktop/Phonepe_Pulse/data/map/transaction/hover/country/india/state/"
state_list = os.listdir(path)

# Create an empty dictionary to store the extracted data
clm = {'State': [], 'Year': [], 'Quarter': [], 'District': [], 'Transaction_count': [], 'Transaction_amount': []}

# Loop over each state's directory to extract transaction data for each district
for i in state_list:
    state_path = path + i + "/"
    year_list = os.listdir(state_path)
    for year in year_list:
        year_path = state_path + year + "/"
        file_list = os.listdir(year_path)
        for filename in file_list:
            file_path = year_path + filename
            with open(file_path, 'r') as f:
                data = json.load(f)
                try:
                    for district_data in data['data']["hoverDataList"]:
                        district = district_data['name']
                        transaction_count = district_data['metric'][0]['count']
                        transaction_amount = district_data['metric'][0]['amount']
                        clm['District'].append(district)
                        clm['Transaction_count'].append(transaction_count)
                        clm['Transaction_amount'].append(transaction_amount)
                        clm['State'].append(i)
                        clm['Year'].append(year)
                        clm['Quarter'].append(int(filename.strip('.json')))
                except:
                    pass 

# Create a pandas DataFrame from the extracted data
map_transaction = pd.DataFrame(clm)
map_transaction.to_csv('district_map_transaction.csv')
map_transaction
```

```python
import os
import json
import pandas as pd

data_path = os.path.join("C:/", "Users", "Sahith", "Desktop", "Phonepe_Pulse", "data", "map", "user", "hover", "country", "india", "state")
state_list = os.listdir(data_path)


# This is to extract the data to create a dataframe

columns = {'State': [], 'Year': [], 'Quarter': [], 'District': [], 'Registered User': [], 'App Opening': []}
for state in state_list:
    state_path = os.path.join(data_path, state)
    years = os.listdir(state_path)
    for year in years:
        year_path = os.path.join(state_path, year)
        files = os.listdir(year_path)
        for file in files:
            file_path = os.path.join(year_path, file)
            with open(file_path, 'r') as f:
                data = json.load(f)
                try:
                    for district in data['data']['hoverData']:
                        registered_user = data['data']['hoverData'][district]['registeredUsers']
                        app_opening = data['data']['hoverData'][district]['appOpens']
                        columns['State'].append(state)
                        columns['Year'].append(year)
                        columns['Quarter'].append(int(file.strip('.json')))
                        columns['District'].append(district)
                        columns['Registered User'].append(registered_user)
                        columns['App Opening'].append(app_opening)
                except:
                    pass       
                 
district_registering = pd.DataFrame(columns)
district_registering.to_csv('district_registering_map.csv')
district_registering
```
